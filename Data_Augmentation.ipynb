```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img("H:\Chrome_downloads\cat\A-Cat.jpg",target_size=(200,200))
```

```python
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
img = image.img_to_array(img)
```

```python
img
```

```python
img.shape
```

```python
input_batch = img.reshape((1,200,200,3))
```

```python
i = 0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir=r"H:\Chrome_downloads\cat\Augmented"):
  i += 1
  if i > 3:
    break
```
